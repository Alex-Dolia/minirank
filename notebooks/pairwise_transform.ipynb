This tutorial introduces the concept of pairwise preference used in most ranking problems. I'll use scikit-learn and [minirank](https://github.com/fabianp/minirank) for learning and matplotlib for visualization.

In the ranking setting, training data consists of lists of items with some order specified between items in each list. This order is typically induced by giving a numerical or ordinal score or a binary judgment (e.g. "relevant" or "not relevant") for each item, so that for any two target values $a, b \in \mathcal{J}$, either $a \prec b$, $b \prec a$ or $b$ and $a$ are not comparable.

More formally, the set of samples $\mathcal{X}$ is embedded with partial order $\prec$. Since in the partial order not all pairs are comparable, the equivalence relationship "a is comparable to b", $a \text{~} b$, generates a partition on $\mathcal{X} = \mathcal{X}_1 \cup \mathcal{X}_2 \cup ... \mathcal{X}_n$, where within block $\mathcal{X}_i$ all elements are comparable. In the classification setting, the class labels are not comparable and thus there are as many parts $S_i$ as elements in the dataset. On the other hand, in regression target values live in a metric space and thus there's a total order between the different target values. In this case, our partition would only have one part, $S = S_1$.

The partition generated by the partial order usually domain-specific. For instance, in information retrieval each block is refered to as a "query id". The goal behind this is to compare only documents that belong to the same query ([Joachims 2002](http://dx.doi.org/10.1145/775047.775067)). In this case, the set of different queries defines the partiton. In medical imaging on the other hand, the order of the labels usually depend on the subject so the partition is given by the different subjects in the study ([Pedregosa et al 2012](http://hal.inria.fr/hal-00717990/en)).

```python
import numpy as np
from scipy import stats
import pylab as pl
from sklearn import linear_model, cross_validation
import minirank
```

To start with, we'll create a dataset in which the target values consists of three graded measurements $\mathcal{Y} = \\{0, 1, 2\\}$ and the input data is a collection of 30 samples, each one with two features. 

The partition generated by the partial order will consist of two equally sized blocks, $\mathcal{X} = \mathcal{X}_1 \cup \mathcal{X}_2$, where each block is generated using a normal distribution with different means. In the pictures, we represent $\mathcal{X}_1$ with round markers and $\mathcal{X}_2$ with triangular markers.

```python
np.random.seed(0)
theta = np.deg2rad(60)
w = np.array([np.sin(theta), np.cos(theta)])
K = 20
X = np.random.randn(K, 2)
y = [0] * K
for i in range(1, 3):
    X = np.concatenate((X, np.random.randn(K, 2) + i * 4 * w))
    y = np.concatenate((y, [i] * K))

# slightly displace data corresponding to our second partition
X[::2] -= np.array([3, 7]) 
blocks = np.array([0, 1] * (X.shape[0] / 2))

# split into train and test set
cv = cross_validation.StratifiedShuffleSplit(y, test_size=.5)
train, test = iter(cv).next()
X_train, y_train, b_train = X[train], y[train], blocks[train]
X_test, y_test, b_test = X[test], y[test], blocks[test]

# plot the result
idx = (b_train == 0)
pl.scatter(X_train[idx, 0], X_train[idx, 1], c=y_train[idx], marker='^', cmap=pl.cm.Blues, s=100)
pl.scatter(X_train[~idx, 0], X_train[~idx, 1], c=y_train[~idx], marker='o', cmap=pl.cm.Blues, s=100)
pl.arrow(0, 0, 8 * w[0], 8 * w[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
pl.text(0, 1, '$w$', fontsize=20)
pl.arrow(-3, -8, 8 * w[0], 8 * w[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
pl.text(-2.6, -7, '$w$', fontsize=20)
pl.axis('equal')
pl.show()
```

The assumption in our case is that for the different partitions in our dataset there's a common vector $w$ such that the projection onto $w$ gives a list with the correct ordering.

If we apply linear regression to the problem above ...

```python
ridge = linear_model.Ridge(1.)
ridge.fit(X_train, y_train)
pl.scatter(X_train[idx, 0], X_train[idx, 1], c=y_train[idx], marker='^', cmap=pl.cm.Blues, s=100)
pl.scatter(X_train[~idx, 0], X_train[~idx, 1], c=y_train[~idx], marker='o', cmap=pl.cm.Blues, s=100)
pl.arrow(0, 0, 20 * ridge.coef_[0], 20 * ridge.coef_[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
pl.text(2, 0, '$\hat{w}$', fontsize=20)
pl.axis('equal')
pl.title('Estimation by Ridge regression')
pl.show()
```

To assess the quality of our model we need a distance. Since we are interesting in a model that *orders* the data, it is natural to look for a metric that compares the ordering of our model to the given ordering. For this, we use [Kendall's tau correlation coefficient](http://en.wikipedia.org/wiki/Kendall_tau_rank_correlation_coefficient), which is defined as 

### $\tau = \frac{P - Q}{P + Q}$ ###

being P is the number of concordant pairs and Q is the number of discordant pairs. This measure is used extensively in the ranking litterature (e.g [Optimizing Search Engines using Clickthrough Data](http://www.cs.cornell.edu/people/tj/publications/joachims_02c.pdf)).

We thus evaluate this metric on the test set for each block separately.

```python
for i in range(2):
    tau, _ = stats.kendalltau(ridge.predict(X_test[b_test == i]), y_test[b_test == i])
    print('Kendall correlation coefficient for block %s: %.5f' % (i, tau))
```

And if the vector $w$ has a perfect ranking score, then the binary-class problem is separable. Not only that, the hyperplane $\\{x^T w = 0\\}$ separates both classes.

We will use this property to learn a classifier in the transformed space.

```python
X_trans, y_trans, diff = minirank.transform_pairwise(X_train, y_train, b_train)
pl.scatter(X_trans[:, 0], X_trans[:, 1], c=diff, s=60, marker='o', cmap=pl.cm.Blues)
pl.arrow(0, 0, 4 * w[0], 4 * w[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
x_space = np.linspace(-10, 10)
pl.plot(x_space * w[1], - x_space * w[0], color='gray')
pl.text(1, 1, '$w$', fontsize=20)
pl.text(3, -4, '$\{x^T w = 0\}$', fontsize=17)
pl.axis('equal')
pl.show()
```

We will now finally train a ranking model bla bla bla

```python
coef_est, _ = minirank.sgd_train((X_train, y_train, b_train), .1, max_iter=50000)
pl.scatter(X_train[idx, 0], X_train[idx, 1], c=y_train[idx], marker='^', cmap=pl.cm.Blues, s=100)
pl.scatter(X_train[~idx, 0], X_train[~idx, 1], c=y_train[~idx], marker='o', cmap=pl.cm.Blues, s=100)
pl.arrow(0, 0, 10 * coef_est[0], 10 * coef_est[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
pl.arrow(-3, -8, 10 * coef_est[0], 10 * coef_est[1], fc='gray', ec='gray', head_width=0.5, head_length=0.5)
pl.text(1, .7, '$\hat{w}$', fontsize=20)
pl.text(-2.6, -7, '$\hat{w}$', fontsize=20)
pl.axis('equal')
pl.show()
```

```python
for i in range(2):
    tau, _ = stats.kendalltau(np.dot(X_test[b_test == i], coef_est), y_test[b_test == i])
    print('Kendall correlation coefficient for block %s: %.5f' % (i, tau))
```

This is indeed higher than the values (0.71122, 0.84387) obtained in the case of linear regression.
